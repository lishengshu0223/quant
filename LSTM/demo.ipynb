```python
path = "F:\Pfactor_clean"
file_name_list = os.listdir(path)
factor_concat = {}
for file_name in tqdm(file_name_list):
    factor_name = file_name[:-4]
    if factor_name[0] != 'a':
        print(factor_name)
        continue
    factor = pd.read_csv(f"{path}\\{file_name}", index_col=0)
    factor.index = pd.to_datetime(factor.index)
    factor = factor.loc["2020-01-01":"2021-01-01",: ]
    factor_concat[factor_name] = factor

factor_concat = pd.concat(factor_concat, axis=1)
factor_concat.to_pickle("F:\\Neural_Networks\data\\factor_concat_2020_2021.pkl")
factor_concat
```

```python
stock_price = pd.read_pickle("F:\Trade_data\\adjopen.pkl")
stock_return = stock_price.pct_change(5).shift(-6)
stock_return.to_pickle("F:\\Neural_Networks\data\stock_return.pkl")
stock_return
```

```python
quantile = 5
quantile_return = stock_return.stack().groupby('dt').apply(
    lambda x: pd.qcut(x, np.arange(quantile + 1) / quantile, np.arange(quantile)))
quantile_return[quantile_return < (quantile - 1)] = 0
quantile_return[quantile_return == (quantile - 1)] = 1
quantile_return
```

```python
from torch.utils.data import DataLoader
from LSTM.dataset import MyDataset
from tqdm import tqdm
import torch

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = pd.read_pickle(".\data\model\model_2018.pkl")
model.to(device)
dataset = MyDataset(factor_concat, stock_return, pd.Timestamp(2018, 1, 1), pd.Timestamp(2019, 12, 31), period=22)
dataloader = DataLoader(dataset, batch_size=1, shuffle=False)
```

```python
score = stock_return.copy()
score.loc[:, :] = np.nan
for code, date, x, y in tqdm(dataloader):
    x = x.float().to(device)
    y_predict = model(x.float())
    score.loc[date, code] = torch.sigmoid(y_predict).cpu().detach().numpy()[0, -1]
```
